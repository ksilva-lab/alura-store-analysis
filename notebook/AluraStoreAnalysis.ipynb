# Importação dos dados



```python
import pandas as pd

url1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja1 = pd.read_csv(url1)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja3
```

#1. Análise do faturamento



```python
#Calculo Faturamento por loja
faturamento_loja1 = loja1['Preço'].sum()
faturamento_loja2 = loja2['Preço'].sum()
faturamento_loja3 = loja3['Preço'].sum()
faturamento_loja4 = loja4['Preço'].sum()
```

```python
#Imprimindo todos os faturamentos
print(f'Total loja 1: R$ {faturamento_loja1:,.2f}')
print(f'Total loja 2: R$ {faturamento_loja2:,.2f}')
print(f'Total loja 3: R$ {faturamento_loja3:,.2f}')
print(f'Total loja 4: R$ {faturamento_loja4:,.2f}')
```

# 2. Vendas por Categoria

```python
#Calculo da quantidade de produtos vendidos por categoria
categoria_loja1 = loja1['Categoria do Produto'].value_counts()
categoria_loja2 = loja2['Categoria do Produto'].value_counts()
categoria_loja3 = loja3['Categoria do Produto'].value_counts()
categoria_loja4 = loja4['Categoria do Produto'].value_counts()
```

```python
#Imprimindo valores calculado por categoria de produtos
print(f'Vendas por categoria Loja 1: {categoria_loja1}\n')
print(f'Vendas por categoria Loja 2: {categoria_loja2}\n')
print(f'Vendas por categoria Loja 3: {categoria_loja3}\n')
print(f'Vendas por categoria Loja 4: {categoria_loja4}\n')
```

# 3. Média de Avaliação das Lojas


```python
# Calculo para a média de avaliação das lojas
avaliacao_loja1 = loja1['Avaliação da compra'].mean()
avaliacao_loja2 = loja2['Avaliação da compra'].mean()
avaliacao_loja3 = loja3['Avaliação da compra'].mean()
avaliacao_loja4 = loja4['Avaliação da compra'].mean()
```

```python
# Resultado das médias calculadas por loja, com arredondamento de 2 casas decimais
print(f'Média de avaliação da loja 1: {avaliacao_loja1:.2f}')
print(f'Média de avaliação da loja 2: {avaliacao_loja2:.2f}')
print(f'Média de avaliação da loja 3: {avaliacao_loja3:.2f}')
print(f'Média de avaliação da loja 4: {avaliacao_loja4:.2f}')
```

# 4. Produtos Mais e Menos Vendidos

## Loja 1

```python
# Lista de todos os produtos vendidos
vendas_loja1 = loja1['Produto'].value_counts()
```

**PRODUTO(S) MAIS VENDIDOS**

```python
# Quantidade do(s) produtos mais vendidos
qtd_max_loja1 = loja1['Produto'].value_counts().max()

# Nome do(s) produto(s) mais vendidos
prod_max_loja1 = vendas_loja1[vendas_loja1 == qtd_max_loja1].index.tolist()

# Apresentação os resultados
print('Produtos mais vendidos na Loja 1')

for produto in prod_max_loja1:
  print(f'- {produto}: {qtd_max_loja1}')
```

**PRODUTO(S) MENOS VENDIDOS**

```python
# Quantidade do(s) produtos menos vendidos
qtd_min_loja1 = loja1['Produto'].value_counts().min()

# Nome do(s) produto(s) menos vendidos
prod_min_loja1 = vendas_loja1[vendas_loja1 == qtd_min_loja1].index.tolist()

# Apresentação dos resultados
print('Produtos menos vendidos na Loja 1')

for produto in prod_min_loja1:
  print(f'- {produto}: {qtd_min_loja1}')
```

# 5. Frete Médio por Loja

```python
media_frete1 = loja1['Frete'].mean()
print(f'O custo médio em frete da primeiro loja é igual a : R$ {media_frete1:.2f} ')

media_frete2 = loja2['Frete'].mean()
print(f'O custo médio em frete da segunda loja é igual a : R$ {media_frete2:.2f} ')

media_frete3 = loja3['Frete'].mean()
print(f'O custo médio em frete da terceira loja é igual a : R$ {media_frete3:.2f} ')

media_frete4 = loja4['Frete'].mean()
print(f'O custo médio em frete da quarta loja é igual a : R$ {media_frete4:.2f} ')
```

```python
import matplotlib.pyplot as plt
import numpy as np

lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']
faturamento = [faturamento_loja1, faturamento_loja2, faturamento_loja3, faturamento_loja4]
avaliacao = [avaliacao_loja1, avaliacao_loja2, avaliacao_loja3, avaliacao_loja4]
frete = [media_frete1, media_frete2, media_frete3, media_frete4]
```

#Relatório de Análise de Desempenho das Lojas – Alura Store


```python
fig, ax1 = plt.subplots(figsize=(10,6))

# barras de faturamento
barras = ax1.bar(lojas, faturamento)

ax1.set_title('Faturamento x Frete Médio por Loja')
ax1.set_xlabel('Lojas')
ax1.set_ylabel('Faturamento (R$)')
ax1.set_yticks([])

# valores acima das barras
for barra in barras:
    valor = barra.get_height()
    ax1.text(
        barra.get_x() + barra.get_width()/2,
        valor,
        f'R$ {valor/1e6:.2f}M',
        ha='center',
        va='bottom',
        fontsize=10
    )

# eixo secundário para frete
ax2 = ax1.twinx()

ax2.plot(lojas, frete, marker='o', linestyle='--', color='red', label='Frete médio')
ax2.set_ylabel('Frete médio (R$)')

# legenda
fig.legend(loc='upper right')

plt.show()
```

O presente relatório tem como objetivo analisar o desempenho das quatro lojas da rede Alura Store para auxiliar o Senhor João na decisão de qual unidade vender para investir em um novo empreendimento.

Para essa análise, foram considerados diferentes indicadores de desempenho, incluindo o faturamento total das lojas, as categorias de produtos mais e menos vendidas, a média de avaliações dos clientes, os produtos mais vendidos e o custo médio de frete. ## Análise do faturamento das lojas

---



O faturamento total é um dos principais indicadores de desempenho de uma loja, pois representa o volume de vendas gerado.

Os valores analisados foram:

- **Loja 1:** R$ 1.534.509,12

- **Loja 2:** R$ 1.488.459,06

- **Loja 3:** R$ 1.464.025,03

- **Loja 4:** R$ 1.384.497,58

A análise do gráfico de faturamento mostra que a **Loja 1 apresenta o maior faturamento**, seguida pela Loja 2 e Loja 3. Já a **Loja 4 apresenta o menor faturamento entre todas**, ficando abaixo da média geral das lojas.

Esse resultado indica que a Loja 4 possui um desempenho financeiro inferior em comparação às demais unidades da rede.

```python
media_faturamento = np.mean(faturamento)

plt.figure(figsize=(10,6))

plt.bar(lojas, faturamento)


# linha da média
plt.axhline(media_faturamento, linestyle='--', color = 'red', label=f'Média: R$ {media_faturamento/1e6:.2f}M')

plt.title('Faturamento Total por Loja')
plt.xlabel('Lojas')
plt.ylabel('Faturamento (R$)')
plt.yticks([])

for barra in barras:
    valor = barra.get_height()
    plt.text(
        barra.get_x() + barra.get_width()/2,
        valor,
        f'R$ {valor/1e6:.2f}M',
        ha='center',
        va='bottom',
        fontsize=10
    )


plt.legend()

plt.show()
```

## Análise das categorias de produtos

---



A análise das vendas por categoria mostrou que algumas categorias concentram grande parte das vendas em todas as lojas. Entre as categorias mais vendidas, destacam-se:

- Móveis

- Eletrônicos

- Brinquedos

Essas categorias apresentam alta demanda e contribuem significativamente para o faturamento das lojas.

Por outro lado, as categorias com menor volume de vendas foram:

- Utilidades domésticas

- Livros

- Instrumentos musicais

De forma geral, todas as lojas apresentam padrões de vendas semelhantes entre as categorias, indicando que o portfólio de produtos possui uma demanda relativamente consistente em toda a rede.

```python
# Criar dataframe com todas as lojas
categorias = pd.DataFrame({
    'Loja 1': categoria_loja1,
    'Loja 2': categoria_loja2,
    'Loja 3': categoria_loja3,
    'Loja 4': categoria_loja4
}).fillna(0)

# Criar coluna com total de vendas por categoria
categorias['Total'] = categorias.sum(axis=1)

# Ordenar categorias pelo total
categorias = categorias.sort_values('Total')

# Remover coluna total para o gráfico
categorias_plot = categorias.drop('Total', axis=1)

# Criar gráfico
categorias_plot.plot(
    kind='barh',
    figsize=(10,6),
    color=['#0d47a1', '#1976d2', '#42a5f5', '#90caf9']
)

plt.title('Vendas por Categoria - Comparação Enttre Lojas')
plt.xlabel('Quantidade de Vendas')
plt.ylabel('')

plt.legend(title='Lojas')

plt.show()
```
